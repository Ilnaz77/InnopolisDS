## Определяем CNN 
### 5. Определить сверточную нейронную сеть
Определите модель CNN, которую можно обучить в наборе данных Fashion-MNIST. Модель должна содержать два сверточных слоя, два пулинга и два полносвязанных слоя. Вы можете использовать любое количество нейронов на слой, если модель принимает изображение размером 28x28 и возвращает результат 10. 

```python
def calc_out_shape_conv(input_matrix_shape, out_channels, kernel_size, stride, padding):
    h_out = np.fix((input_matrix_shape[2]+2*padding-kernel_size)/stride + 1)
    w_out = np.fix((input_matrix_shape[3]+2*padding-kernel_size)/stride + 1)
    out_shape = [input_matrix_shape[0],out_channels,h_out,w_out]
    return out_shape

def calc_out_shape_pooling(input_matrix_shape, pool_size, stride, padding):
    output_maxtrix_shape = input_matrix_shape
    output_maxtrix_shape[2] = (input_matrix_shape[2] - pool_size + 2*padding)/stride + 1
    output_maxtrix_shape[3] = (input_matrix_shape[3] - pool_size + 2*padding)/stride + 1
    return output_maxtrix_shape

print(calc_out_shape([10, 1, 28, 28], 8, 5, 1, 2))
print(calc_out_shape_pooling([10, 8, 28.0, 28.0], 2, 2, 0))

print(calc_out_shape([10, 8, 14.0, 14.0], 16, 5, 1, 0))
print(calc_out_shape_pooling([10, 16, 10.0, 10.0], 2, 2, 0))
```

```python
class ConvolutionalNetwork(nn.Module):
    def __init__(self, channels=8, kernel_size=5):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=channels, kernel_size=kernel_size, padding=2)
        self.act1 = nn.Tanh()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        self.conv2 = nn.Conv2d(in_channels=channels, out_channels=2*channels, kernel_size=kernel_size, padding=0)
        self.act2 = nn.Tanh()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        self.fc1 = nn.Linear(in_features=16*5*5, out_features=120)
        self.act3 = nn.Tanh()
        self.fc2 = nn.Linear(in_features=120, out_features=10)
        self.act4 = nn.Softmax(dim=1)

    def forward(self, X):
        X = self.conv1(X)
        X = self.act1(X)
        X = self.pool1(X)
        X = self.conv2(X)
        X = self.act2(X)
        X = self.pool2(X)
        X = X.view(-1, 16*5*5) # спросить почему не юзаем reshape?
        X = self.fc1(X)
        X = self.act3(X)
        X = self.fc2(X)
        return X
    def predict_proba(self, X):
        X = self.forward(X)
        X = self.act4(X)
        return X
        
model = ConvolutionalNetwork()
```

### 7. Определить функцию потерь и оптимизатор

```python
# ВАШ КОД
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

### 8. Тренируем модель
Обучайте модель 5 эпох. <br>
ДОПОЛНИТЕЛЬНО: Сделайте вывод после каждой эпохи, чтобы показать прогресс в обучении. Измерьте время обучения на одну эпоху.

```python
# ВАШ КОД
# спросить про forward и eval и torch.no_grad()
# почему юзаем model(X_train) вместо model.forward(X_train)
# почему считаем предикты и лосс по батчам потом их складываем и усредняем? Это делается только п., что памяти может не хватить? 
epochs = 10

test_loss_history = []
train_loss_history = []

test_accuracy_history = []
train_accuracy_history = []

batch_train_loss_history = []
batch_train_accuracy_history = []

amount_train = 60000.0
amount_test = 10000.0


for epoch in range(1, epochs+1):
    sum_pre_accur_on_train_batch = 0
    sum_pre_accur_on_test_batch = 0
    sum_loss_on_train_batch = 0
    sum_loss_on_test_batch = 0
    
    start_time = time.time()
    model.train() # включаем режим обучения
    # итерация по бачам
    for X_batch_train, y_batch_train in train_loader:
        optimizer.zero_grad()
        y_predict = model.forward(X_batch_train)
        
        pre_accur_on_batch = (y_predict.argmax(dim=1) == y_batch_train).float().sum()
        sum_pre_accur_on_train_batch += pre_accur_on_batch
        
        loss = criterion(y_predict, y_batch_train)
        sum_loss_on_train_batch += loss.item()
        loss.backward()
        optimizer.step()
    print(f'Время выполнения одной эпохи: {time.time() - start_time:.0f} seconds')
    model.eval() # выключиаем режим обучения
    with torch.no_grad():
        for X_batch_test, y_batch_test in test_loader:
            y_validation = model.forward(X_batch_test)
            pre_accur_on_batch = (y_validation.argmax(dim=1) == y_batch_test).float().sum()
            sum_pre_accur_on_test_batch += pre_accur_on_batch
            loss = criterion(y_validation, y_batch_test)
            sum_loss_on_test_batch += loss.item()
    test_accuracy_history.append(sum_pre_accur_on_test_batch.item()*100/amount_test)
    train_accuracy_history.append(sum_pre_accur_on_train_batch.item()*100/amount_train)
    test_loss_history.append(sum_loss_on_test_batch/(amount_test/10))
    train_loss_history.append(sum_loss_on_train_batch/(amount_train/10))
    print('epoch: ', epoch, 'accur_train: ', np.round(sum_pre_accur_on_train_batch.item()*100/amount_train, 2), 'accur_test: ', np.round(sum_pre_accur_on_test_batch.item()*100/amount_test, 2))
```

```python
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(1,2,1)
ax1.plot(test_accuracy_history, marker='o', label='test')
ax1.plot(train_accuracy_history, marker='o', label='train')
ax1.set(title='Accuracy', xlabel='Epoch', ylabel='Accuracy%')
ax1.legend()

ax2 = fig.add_subplot(1,2,2)
ax2.plot(test_loss_history, label='test', marker='o')
ax2.plot(train_loss_history, label='train', marker='o')
ax2.set(title='Loss', xlabel='Epoch', ylabel='CrossEntropy')
ax2.legend()

plt.show()
```

```python
def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    df_cm = pd.DataFrame(confusion_matrix, index=class_names, columns=class_names)
    fig = plt.figure(figsize=figsize)
    heatmap = sns.heatmap(df_cm, annot=True, fmt="d", )
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.xlabel('True label')
    plt.ylabel('Predicted label')
    
print_confusion_matrix(confusion_matrix(y_batch_test.view(-1), y_predict.argmax(dim=1).view(-1)).T, class_names)
```
